- If the dataset is properly formatted as in the above dataset it is simple to use the mean() function that is built in pandas after grouping the data by ASN and Co server respectively, as shown below, from here it is easy to find the avg throughput of any required ASN or Co Server

```python
group_ASN = data.groupby("ASN").mean().Bytes
group_ASN
```

```python
group_CoServer = data.groupby("CoServer").mean().Bytes
group_CoServer
```
